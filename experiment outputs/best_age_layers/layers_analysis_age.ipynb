```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_csv('/content/age_layers/data_eval_age_final_4_0.001_512_1.csv')
df_2 = pd.read_csv('/content/age_layers/data_eval_age_final_4_0.001_512_2.csv')
df_3 = pd.read_csv('/content/age_layers/data_eval_age_final_4_0.001_512_3.csv')
```

```python
def df_to_list(df):
    data2=[]
    for i in range(len(df)):
        data2.append(df['val_accuracy'][i])
    return data2
```

```python
df_1_l=[]
df_2_l=[]
df_3_l=[]
df_1_l=df_to_list(df_1)
df_2_l=df_to_list(df_2)
df_3_l=df_to_list(df_3)
```

```python
df_list=[]
df_list.append(df_1_l)
df_list.append(df_2_l)
df_list.append(df_3_l)
```

```python
num_layers=[1,2,3]
plt.figure(2)
#plt.plot(range(len(neuron_numbers)), neuron_acc)
#plt.xticks(range(len(neuron_numbers)), neuron_numbers)
for l in range(3):
    plt.plot(range(len(df_list[l])), df_list[l], label = 'layers = {}'.format(num_layers[l]))
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('cross-validation accuracies against training epochs for different numbers of hidden-layer neurons')
plt.legend()
```
