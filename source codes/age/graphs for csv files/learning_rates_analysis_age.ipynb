```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_01 = pd.read_csv('/Users/abhi/Desktop/NNDL Project/results/age_learning/data_eval_age_final_4_0.01_512_2.csv')
df_001 = pd.read_csv('/Users/abhi/Desktop/NNDL Project/results/age_learning/data_eval_age_final_4_0.001_512_2.csv')
df_005 = pd.read_csv('/Users/abhi/Desktop/NNDL Project/results/age_learning/data_eval_age_final_4_0.005_512_2.csv')
```

```python
def df_to_list(df):
    data2=[]
    for i in range(len(df)):
        data2.append(df['val_accuracy'][i])
    return data2
```

```python
df_01_l=[]
df_001_l=[]
df_005_l=[]
df_01_l=df_to_list(df_01)
df_001_l=df_to_list(df_001)
df_005_l=df_to_list(df_005)
```

```python
df_list=[]
df_list.append(df_01_l)
df_list.append(df_001_l)
df_list.append(df_005_l)
```

```python
num_neurons=[0.01,0.001,0.005]
plt.figure(2)
#plt.plot(range(len(neuron_numbers)), neuron_acc)
#plt.xticks(range(len(neuron_numbers)), neuron_numbers)
for l in range(3):
    plt.plot(range(len(df_list[l])), df_list[l], label = 'layers = {}'.format(num_neurons[l]))
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('cross-validation accuracies against training epochs for different numbers of hidden-layer neurons')
plt.legend()
```
