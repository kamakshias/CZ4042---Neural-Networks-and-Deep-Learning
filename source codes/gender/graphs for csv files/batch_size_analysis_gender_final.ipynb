```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_2 = pd.read_csv('/content/outputs_age_batch/data_eval_gender_final_2_0.001_512_4.csv')
df_4 = pd.read_csv('/content/outputs_age_batch/data_eval_gender_final_4_0.001_512_4.csv')
df_8 = pd.read_csv('/content/outputs_age_batch/data_eval_gender_final_8_0.001_512_4.csv')
df_16 = pd.read_csv('/content/outputs_age_batch/data_eval_gender_final_16_0.001_512_4.csv')
df_32 = pd.read_csv('/content/outputs_age_batch/data_eval_gender_final_32_0.001_512_4.csv')
df_64 = pd.read_csv('/content/outputs_age_batch/data_eval_gender_final_64_0.001_512_4.csv')
```

```python
def df_to_list(df):
    data2=[]
    for i in range(len(df)):
        data2.append(df['val_accuracy'][i])
    return data2
```

```python
df_2_l=[]
df_4_l=[]
df_8_l=[]
df_16_l=[]
df_32_l=[]
df_64_l=[]
df_2_l=df_to_list(df_2)
df_4_l=df_to_list(df_4)
df_8_l=df_to_list(df_8)
df_16_l=df_to_list(df_16)
df_32_l=df_to_list(df_32)
df_64_l=df_to_list(df_64)
```

```python
df_list=[]
df_list.append(df_2_l)
df_list.append(df_4_l)
df_list.append(df_8_l)
df_list.append(df_16_l)
df_list.append(df_32_l)
df_list.append(df_64_l)
```

```python
batch_size=[2,4,8,16,32,64]
plt.figure(2)
#plt.plot(range(len(neuron_numbers)), neuron_acc)
#plt.xticks(range(len(neuron_numbers)), neuron_numbers)
for l in range(6):
    plt.plot(range(len(df_list[l])), df_list[l], label = 'batch_size = {}'.format(batch_size[l]))
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('cross-validation accuracies against training epochs for different numbers of hidden-layer neurons')
plt.legend()
```
