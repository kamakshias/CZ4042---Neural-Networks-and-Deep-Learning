```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_256 = pd.read_csv('/content/age_neurons_analysis/data_eval_age_final_4_0.001_256_4.csv')
df_512 = pd.read_csv('/content/age_neurons_analysis/data_eval_age_final_4_0.001_512_4.csv')
df_1024 = pd.read_csv('/content/age_neurons_analysis/data_eval_age_final_4_0.001_1024_4.csv')
```

```python
def df_to_list(df):
    data2=[]
    for i in range(len(df)):
        data2.append(df['val_accuracy'][i])
    return data2
```

```python
df_256_l=[]
df_512_l=[]
df_1024_l=[]
df_256_l=df_to_list(df_256)
df_512_l=df_to_list(df_512)
df_1024_l=df_to_list(df_1024)
```

```python
df_list=[]
df_list.append(df_256_l)
df_list.append(df_512_l)
df_list.append(df_1024_l)
```

```python
num_neurons=[256,512,1024]
plt.figure(2)
#plt.plot(range(len(neuron_numbers)), neuron_acc)
#plt.xticks(range(len(neuron_numbers)), neuron_numbers)
for l in range(3):
    plt.plot(range(len(df_list[l])), df_list[l], label = 'layers = {}'.format(num_neurons[l]))
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('cross-validation accuracies against training epochs for different numbers of hidden-layer neurons')
plt.legend()
```
